```python
# we have to remove M and k
def clean_size(index):
  if str(index)[-1]=="M":
    res=float(str(index).replace("M",""))
    res=res*1024
    return res
  elif(str(index)[-1]=="k"):
    res=float(str(index).replace("k",""))
    return res
  else:
    return np.NaN
```

```python
s="19k"
clean_size(s)
```

```python
#this is the way you can apply to function
df["Size"]=df["Size"].apply(clean_size)
df.Size
```
